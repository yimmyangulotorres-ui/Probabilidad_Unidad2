# Tarea 1: Modelado Computacional de la Distribución Binomial

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n_ensayos = 20
p_falla = 0.15

x = np.arange(0, n_ensayos + 1)

pmf_binomial = binom.pmf(x, n_ensayos, p_falla)
cdf_binomial = binom.cdf(x, n_ensayos, p_falla)

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

ax[0].vlines(x, 0, pmf_binomial, colors='b', lw=5, alpha=0.5)
ax[0].plot(x, pmf_binomial, 'bo', ms=8)
ax[0].set_title(f'PMF - Binomial (n={n_ensayos}, p={p_falla})')
ax[0].set_xlabel('Número de fallos (x)')
ax[0].set_ylabel('P(X = x)')
ax[0].grid(True, alpha=0.3)

ax[1].step(x, cdf_binomial, where='post', color='r', lw=2)
ax[1].plot(x, cdf_binomial, 'ro', alpha=0.5)
ax[1].set_title('CDF - Distribución Acumulada')
ax[1].set_xlabel('Número de fallos (x)')
ax[1].set_ylabel('P(X <= x)')
ax[1].grid(True, alpha=0.3)

plt.show()
plt.tight_layout()
plt.show()

# Cálculo de probabilidad específica: P(X <= 3)
prob_max_3 = binom.cdf(3, n_ensayos, p_falla)
print(f"La probabilidad de tener 3 fallos o menos es: {prob_max_3:.4f}")
```

# **Tarea 2: Modelado de la Distribución de Poisson (Eventos Raros)**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 1. Definir los parámetros dados en el problema
lmbda = 4.5 # Tasa media de peticiones erróneas por minuto
x_min = 0
x_max = 15

# 2. Generar el rango de valores para x (de 0 a 15)
x = np.arange(x_min, x_max + 1)

# 3. Calcular la Función de Masa de Probabilidad (PMF) para cada valor de x
pmf_valores = poisson.pmf(x, lmbda)

# 4. Graficar la PMF
plt.figure(figsize=(10, 6))
# Usamos un gráfico de tallos (stem) o líneas verticales que es ideal para distribuciones discretas
plt.vlines(x, 0, pmf_valores, colors='royalblue', lw=4, alpha=0.7)
plt.plot(x, pmf_valores, 'bo', ms=8)

# Configuración estética del gráfico
plt.title('PMF - Distribución de Poisson ($\lambda = 4.5$)', fontsize=14)
plt.xlabel('Número de peticiones erróneas ($x$)', fontsize=12)
plt.ylabel('Probabilidad $P(X=x)$', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar la gráfica
plt.show()

# 5. Calcular la probabilidad exacta de recibir exactamente 6 peticiones erróneas P(X=6)
probabilidad_6 = poisson.pmf(6, lmbda)

# Imprimir el resultado
print(f"La probabilidad exacta de recibir exactamente 6 peticiones erróneas en un minuto P(X=6) es: {probabilidad_6:.4f} (o {probabilidad_6 * 100:.2f}%)")
```

# **Tarea 3: Hito del Proyecto - Identificación de Variables de Conteo (ABP)**

```python
# ==========================================
# TAREA 3 - IDENTIFICACIÓN DE VARIABLES DE CONTEO
# Dataset: Datos_Dengue_MSP_Ene2021_Ago2025.xlsx
# ==========================================

# ==========================================
# IMPORTAR LIBRERÍAS
# ==========================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import poisson

# ==========================================
# ESTILO GENERAL
# ==========================================

plt.style.use('ggplot')
sns.set_theme(style="whitegrid")

# ==========================================
# CARGA DEL DATASET
# ==========================================

df_regional = pd.read_excel("Datos_Dengue_MSP_Ene2021_Ago2025.xlsx")

# Mostrar primeras filas
print(df_regional.head())

# ==========================================
# LIMPIAR NOMBRES DE COLUMNAS
# ==========================================

df_regional.columns = df_regional.columns.str.strip()

# ==========================================
# VARIABLE DE CONTEO
# ==========================================

# Se selecciona la variable "Total"
# porque representa el número de casos registrados

variable = 'Total'

# ==========================================
# LIMPIEZA DE DATOS
# ==========================================

df_regional[variable] = pd.to_numeric(
    df_regional[variable],
    errors='coerce'
)

# Eliminar valores nulos
datos = df_regional[variable].dropna()

# ==========================================
# CÁLCULO DE LA MEDIA MUESTRAL
# ==========================================

lambda_poisson = datos.mean()

# Varianza muestral
varianza = datos.var(ddof=1)

# ==========================================
# RESULTADOS
# ==========================================

print("="*50)
print(" ANÁLISIS ESTADÍSTICO DEL DATASET ")
print("="*50)

print(f"\nVariable Analizada: {variable}")
print(f"Media Muestral (λ): {lambda_poisson:.2f}")
print(f"Varianza Muestral: {varianza:.2f}")

# ==========================================
# DISTRIBUCIÓN DE POISSON TEÓRICA
# ==========================================

# Valores del eje X
x = np.arange(0, datos.max() + 1)

# PMF de Poisson
pmf = poisson.pmf(x, lambda_poisson)

# ==========================================
# HISTOGRAMA + PMF POISSON
# ==========================================

plt.figure(figsize=(12,6))

# Histograma de densidad de datos reales
plt.hist(
    datos,
    bins=20,
    density=True,
    alpha=0.7,
    color='skyblue',
    edgecolor='black',
    label='Datos reales'
)

# PMF teórica de Poisson
plt.plot(
    x,
    pmf,
    marker='o',
    linestyle='-',
    color='red',
    linewidth=2,
    label='Poisson teórica'
)

# Línea de la media
plt.axvline(
    lambda_poisson,
    color='blue',
    linestyle='--',
    linewidth=2,
    label=f'λ = {lambda_poisson:.2f}'
)

# ==========================================
# PERSONALIZACIÓN DEL GRÁFICO
# ==========================================

plt.title(
    'Distribución de Casos de Dengue vs Poisson',
    fontsize=16,
    fontweight='bold'
)

plt.xlabel('Número de Casos')
plt.ylabel('Densidad / Probabilidad')

plt.legend()

plt.tight_layout()

plt.show()

# ==========================================
# CONCLUSIÓN
# ==========================================

print("\nCONCLUSIÓN:")
print("La variable 'Total' representa una variable discreta de conteo.")
print(f"La media muestral obtenida fue λ = {lambda_poisson:.2f}.")
print("Al comparar el histograma de los datos reales con la")
print("distribución de Poisson teórica, se observa cierta")
print("similitud general entre ambas distribuciones.")
print("Sin embargo, existen diferencias debido a la dispersión")
print("y a la presencia de valores altos en algunos registros.")
```

# **Tarea 4: ABI - Aproximación Binomial a Poisson**

## Aproximación de la Distribución Binomial a Poisson

La relación matemática:

$$
\lambda = n \cdot p
$$

cuando:

$$
n \to \infty
$$

y

$$
p \to 0
$$

define la aproximación de la distribución Binomial por la distribución de Poisson.

Esta aproximación ocurre cuando el número de ensayos \( n \) es muy grande y la probabilidad de éxito \( p \) es muy pequeña, manteniendo constante el valor de \( \lambda \).

En este contexto:

- \( n \) representa el número de ensayos.
- \( p \) representa la probabilidad de éxito en cada ensayo.
- \( \lambda \) representa el número promedio de ocurrencias de un evento en un intervalo de tiempo o espacio.

Bajo estas condiciones, la distribución:

$$
Binomial(n,p)
$$

se aproxima a:

$$
Poisson(\lambda)
$$

La distribución de Poisson se utiliza para modelar eventos raros e independientes, como:

- fallas en sistemas,
- accidentes,
- llamadas telefónicas,
- errores de transmisión,
- llegada de clientes.

La aproximación es útil porque simplifica los cálculos probabilísticos cuando trabajar directamente con la distribución Binomial resulta complejo debido al gran valor de \( n \).

En conclusión, cuando existen muchos ensayos y la probabilidad de éxito es muy pequeña, la distribución Binomial converge hacia una distribución de Poisson, facilitando el análisis y modelado de fenómenos reales.

## Script en Python

```python
from scipy.stats import binom, poisson

# Parámetros
n = 1000
p = 0.003
x = 2

# Parámetro lambda
lam = n * p

# Probabilidad Binomial
prob_binomial = binom.pmf(x, n, p)

# Probabilidad Poisson
prob_poisson = poisson.pmf(x, lam)

# Resultados
print("Distribución Binomial")
print(f"P(X = {x}) = {prob_binomial:.10f}")

print("\nDistribución Poisson")
print(f"P(X = {x}) = {prob_poisson:.10f}")

# Diferencia absoluta
diferencia = abs(prob_binomial - prob_poisson)

print("\nDiferencia absoluta:")
print(f"{diferencia:.10f}")
```

## **•	Analizando los supuestos del experimento de Bernoulli, si extraemos cartas de una baraja sin reemplazo buscando ases, ¿podemos modelar este escenario con una distribución Binomial? Justifique estadísticamente su respuesta**


## Análisis del experimento de Bernoulli

Analizando los supuestos del experimento de Bernoulli, si extraemos cartas de una baraja **sin reemplazo** buscando ases, ¿podemos modelar este escenario con una distribución Binomial? Justifique estadísticamente su respuesta.

### Respuesta

No es correcto modelar este escenario mediante una **Distribución Binomial** cuando las cartas se extraen **sin reemplazo**.

La distribución binomial se utiliza cuando se cumplen las siguientes condiciones:

1. Existe un número fijo de ensayos.
2. Cada ensayo tiene únicamente dos posibles resultados:
   - Éxito: obtener un as.
   - Fracaso: no obtener un as.
3. La probabilidad de éxito es constante en todos los ensayos.
4. Los ensayos son independientes entre sí.

Aunque en este problema sí existen dos posibles resultados y puede fijarse un número de extracciones, los supuestos de **probabilidad constante** e **independencia** no se cumplen.

### Ejemplo estadístico

Supongamos que se extraen cartas de una baraja de 52 cartas buscando ases.

#### Primera extracción

La probabilidad de obtener un as es:

$$
P(A_1)=\frac{4}{52}
$$

porque existen 4 ases en una baraja de 52 cartas.

---

#### Segunda extracción (sin reemplazo)

Si en la primera extracción salió un as, entonces quedan:

- 3 ases
- 51 cartas totales

La nueva probabilidad sería:

$$
P(A_2 \mid A_1)=\frac{3}{51}
$$

En cambio, si en la primera extracción **no** salió un as, entonces quedan:

- 4 ases
- 51 cartas

y la probabilidad cambia a:

$$
P(A_2 \mid A_1^c)=\frac{4}{51}
$$

---

### Interpretación

Las probabilidades cambian dependiendo del resultado anterior, lo que demuestra que:

- Los ensayos **no son independientes**.
- La probabilidad de éxito **no permanece constante**.

Por esta razón, el experimento no puede modelarse correctamente mediante una distribución binomial.

### Distribución adecuada

La distribución más apropiada para este escenario es la **Distribución Hipergeométrica**, ya que describe experimentos de extracción **sin reemplazo** en poblaciones finitas.
